```python
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings("ignore")
  
MSCIA = pd.read_csv('./MSCIA number.csv')
print('The shape of the dataset is:', MSCIA.shape)
```

```python
MSCIA.head(5)
```

```python
list_of_date = MSCIA['Time'].tolist()
industry_name = list(MSCIA.columns)
industry_name.remove("Time")
```

```python
MSCIA_Modified = MSCIA.drop(MSCIA.columns[0,],axis=1) 
MSCIA_Modified.head(5)
```

```python
MSCIA_NUMBER_table = pd.DataFrame(columns=industry_name, index = list_of_date)
```

```python
print(MSCIA_Modified.shape[0])
row_length = range(MSCIA_Modified.shape[0]-1)
column_length = range(MSCIA_Modified.shape[1])
```

```python
for n in column_length:
    MSCIA_NUMBER_table.loc[list_of_date[375], industry_name[n]] = 1
    k = 1
    for i in row_length:
        rate_day = ((float(MSCIA_Modified.iat[375-(i+1),n])-float(MSCIA_Modified.iat[375-i,n]))/
                           float(MSCIA_Modified.iat[375-i,n])) + 1
        k = k*rate_day
        MSCIA_NUMBER_table.loc[list_of_date[375-i-1], industry_name[n]] = k
```

```python
MSCIA_NUMBER_table.to_csv('MSCIA_NUMBER_table.csv')
```

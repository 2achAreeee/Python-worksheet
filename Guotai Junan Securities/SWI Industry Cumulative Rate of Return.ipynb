```python
Profit_rate_table = pd.DataFrame(columns=data_code, index = list_of_date)

    
```

```python
for n in column_length:
    rate = 1
    Profit_rate_table.loc[list_of_date[374], data_code[n]] = rate
    k = 1
    for i in row_length:
        profit_rate_day = ((float(SWI.iat[374-(i+1),n])-float(SWI.iat[374-i,n]))/float(SWI.iat[374-i,n])) + 1
        k = k*profit_rate_day
        rate = k-1
        Profit_rate_table.loc[list_of_date[374-i-1], data_code[n]] = rate
    
```
